```python
import pandas as pd

# Load the CSV file into a DataFrame
df_house = pd.read_csv('house.csv')
# Group by the StateName column and compute the average for each of the data columns
df_state_avg = df_house.groupby('StateName').mean().drop(columns=['RegionID', 'SizeRank'])

# Reset index for better readability
df_state_avg.reset_index(inplace=True)

df_state_avg.head()
```
